```python
#excercise:
# Create a unit vector at 45 degrees, cos 45 & sin 45 is equal to 0.707
# with 45 degrees we can simulate qubit superposition, and with the degrees we can get the probability of each state
v = np.array([np.sqrt(0.5),np.sqrt(0.5)])
```

```python
# verify that's a unit vector
magnitude = np.linalg.norm(v)
magnitude
```

```python
# create a dictionary with the necessary data to include in our data frame
datv = {
    "X component (State 0)" :[v[0]],
    "Y component (State 1)" :[v[1]],
    "Vector Magnitude" : [magnitude],
    "State 0 probability" : [v[0]**2],
    "State 1 probability": [v[1]**2]
}
# then create the data frame using pandas and print it
tb = pd.DataFrame(datv)
print(tb)
```

```python
fig, ax = plt.subplots() # we created the visualization into a plot

ax.set_xlim(0,1.5)# set boundaries to improve the visualization and create space around our data
ax.set_ylim(0,1.5)

origin_x=np.array([0])
origin_y=np.array([0])
u=np.array(v[0])
w=np.array(v[1])


ax.quiver(origin_x,origin_y,u,w,angles='xy', scale_units='xy',scale=1,color="purple", label='qubit superposition')
# Plot the vector using arrow visualization. 'angles' and 'scale_units' ensure the arrow follows the data coordinates accurately,
#not the screen pixels.

ax.grid(True) # Enable the background grid for better readability.
ax.set_aspect('equal') # Force the X and Y axes to have the same scale (square ratio) to prevent vector distortion.
ax.legend()# Display the legend with the vector label.
plt.title("Quibiting")
plt.show()# Set the plot title and render the final figure.
```
