```python
"""We will simulate and visualize how a rotation matrix affects a qubit.
 We will demonstrate graphically that a unitary matrix rotates the vector without changing its size"""
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


"""First stage:
for the engine to this project we need to define 3 variables:
1. Create the initial state --> vector ∣0⟩
"""
initial = np.array([[1],[0]])
```

```python
# 2. the variable to the rotation angle(the value in rad)
theta = np.pi /4
```

```python
# 3. Transformation matrix
coseno = np.cos(theta)
seno = np.sin(theta)
t_matrix= np.array([[coseno,-seno],[seno,coseno]])
```

```python
#get the final state

final_state =  np.dot(t_matrix,initial)
final_state
```

```python
#Determinant must be equal to 1
det = np.linalg.det(t_matrix)
det
```

```python
"""Stage 2: data report
"""
report = {
    "amplitude X": [initial.item(0),final_state.item(0)],
    "amplitude Y": [initial.item(1),final_state.item(1)],
    "probability":[initial.item(0) **2 + initial.item(1) ** 2, final_state.item(0) **2 + final_state.item(1) ** 2,]
}

report_table = pd.DataFrame(report, index=["Initial State", 'Final State'] )
report_table
```

```python
#Stage 3: Visualization
# We use "angles='xy',scale_units='xy'scale=1" to real scale 1:1
#angles='xy' to ensure the vector begins from the origin
# scale_units='xy' to ensure the unit of scale will be the same as the units of the plot in coordinates x and y.
#scale=1 , scale factor, a vector component that it value is one corresponds to the one unit of the axis


plt.quiver(0,0,initial.item(0),initial.item(1),angles="xy",scale_units="xy", scale=1,color="blue")
plt.quiver(0,0,final_state.item(0),final_state.item(1),angles="xy",scale_units="xy", scale=1,color="red")
plt.xlim(-0.05,1.5)
plt.ylim(-0.05,1.5)
plt.grid()
plt.gca().set_aspect('equal')
plt.show()
```
